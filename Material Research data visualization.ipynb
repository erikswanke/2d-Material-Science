```python
import pandas as pd
import csv
from pandas import read_csv
import matplotlib.pyplot as plt


data = pd.read_csv('magneticmoment_Ef_data.csv', header=0)  

mode = ["add", "sub", "mul", "div"] #supported features at the moment: add, sub, mul, div
feature1 = "energy"
feature2 = "formation_energy"
begindex = 0
endex = 30

for i in mode:
    dt1 = data[feature1].iloc[begindex:endex]
    dt2 = data[feature2].iloc[begindex:endex]
    if(i=="add"):
        dt = dt1.add(dt2)
        op = "+"
    elif(i=="sub"):
        dt = dt1.sub(dt2)
        op = "-"
    elif(i=="mul"):
        dt = dt1.mul(dt2)
        op = "*"
    elif(i=="div"):
        dt = dt1.div(dt2)
        op = "/"

    dt.plot()
    plt.xlabel("element num")
    plt.ylabel(feature1 + " " + op + " " + feature2)
    plt.show()


```
