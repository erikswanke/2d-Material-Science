```python
import pandas as pd
import csv
from pandas import read_csv
import matplotlib.pyplot as plt


data = pd.read_csv('magneticmoment_Ef_data.csv', header=0)  

mode = ["add", "sub", "mul", "div"] #supported features at the moment: add, sub, mul, div
feature1 = "energy"
feature2 = "formation_energy"
begindex = 0
endex = 30


df_list = []
op_list = []
for i in mode:
    dt1 = data[feature1].iloc[begindex:endex]
    dt2 = data[feature2].iloc[begindex:endex]
    if(i=="add"):
        dt = dt1.add(dt2)
        df_list.append(dt)
        op_list.append(" + ")
        
    elif(i=="sub"):
        dt = dt1.sub(dt2)
        df_list.append(dt)
        op_list.append(" - ")
    elif(i=="mul"):
        dt = dt1.mul(dt2)
        df_list.append(dt)
        op_list.append(" * ")
    elif(i=="div"):
        dt = dt1.div(dt2)
        df_list.append(dt)
        op_list.append(" / ")

fig, axes = plt.subplots(2, 2,figsize=(12, 9))
count=0
for r in range(2):
    for c in range(2):
        axes[r][c].set_ylabel(feature1 + op_list[count]+feature2)
        axes[r][c].set_xlabel("Element Number")
        axes[r][c].fmt = '[g]'
        df_list[count].plot(ax=axes[r,c])
        
        count+=1
```
